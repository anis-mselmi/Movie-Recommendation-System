```python
from transformers import pipeline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sentiment_analyzer = pipeline("sentiment-analysis")
```

```python
reviews = [
    "The tacos were absolutely delicious! 🌮😋",
    "The service was very slow and the food was cold.",
    "I loved the pizza, it was fresh and cheesy! 🍕",
    "The place was dirty and I will never come back.",
    "Amazing burgers! Fast service and friendly staff. 🍔",
    "The sushi was okay, but nothing special.",
    "Terrible coffee, worst I’ve ever had.",
    "The pasta was cooked perfectly, I’ll definitely return! 🍝"
]
```

```python
results = []
for review in reviews:
    result = sentiment_analyzer(review)[0]
    results.append(result)
    print(f"Review: {review}")
    print(f"Sentiment: {result['label']} (score: {result['score']:.2f})")
    print("-" * 50)
```

```python
labels = [r["label"] for r in results]
positive_count = labels.count("POSITIVE")
negative_count = labels.count("NEGATIVE")

sns.set_style("whitegrid")
plt.figure(figsize=(6,4))

sns.barplot(x=["Positive", "Negative"], y=[positive_count, negative_count], palette="coolwarm")
plt.title("🍴 Restaurant Review Sentiment Analysis")
plt.ylabel("Number of Reviews")
plt.show()
```
